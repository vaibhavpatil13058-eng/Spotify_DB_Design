```python
num = int(input("Enter a number: "))

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
else:
    print("Not a Prime Number")
```

```python
num = int(input("Enter a number: "))
temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum += digit ** 3
    temp //= 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
# ---------------- BACKEND (Logic) ---------------- #

# Question database
questions = [
    {
        "question": "What is the capital of India?",
        "options": ["A. Mumbai", "B. Delhi", "C. Kolkata", "D. Chennai"],
        "answer": "B"
    },
    {
        "question": "Which language is used for web apps?",
        "options": ["A. Python", "B. Java", "C. JavaScript", "D. All"],
        "answer": "D"
    },
    {
        "question": "What is 5 + 3?",
        "options": ["A. 5", "B. 8", "C. 10", "D. 15"],
        "answer": "B"
    }
]

def check_answer(user_ans, correct_ans):
    return user_ans.upper() == correct_ans


# ---------------- FRONTEND (User Interaction) ---------------- #

def run_quiz():
    score = 0

    print("🎯 Welcome to MCQ Quiz Game\n")

    for i, q in enumerate(questions):
        print(f"Q{i+1}. {q['question']}")

        for option in q["options"]:
            print(option)

        user_ans = input("Enter your answer (A/B/C/D): ")

        if check_answer(user_ans, q["answer"]):
            print("✅ Correct!\n")
            score += 1
        else:
            print(f" Wrong! Correct answer is {q['answer']}\n")

    print("📊 Quiz Completed!")
    print("Your Score:", score, "/", len(questions))


# ---------------- MAIN PROGRAM ---------------- #

run_quiz()
```

```python
password = input("Enter your password: ")

length = len(password)

has_upper = False
has_lower = False
has_digit = False
has_special = False

for char in password:
    if char.isupper():
        has_upper = True
    elif char.islower():
        has_lower = True
    elif char.isdigit():
        has_digit = True
    else:
        has_special = True

# Checking strength
if length >= 8 and has_upper and has_lower and has_digit and has_special:
    print("Strong Password ")
elif length >= 6 and (has_upper or has_lower) and has_digit:
    print("Medium Password ")
else:
    print("Weak Password ")
```
